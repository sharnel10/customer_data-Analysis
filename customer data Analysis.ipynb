### Task #1: Cleaning the data

```python
customer_data = customer_data.drop(columns = ['Gender'])
```

```python
customer_data
```

```python
customer_data['Month'] = customer_data['Date'].str[5:7]
customer_data['Month'] = customer_data['Month'].astype('int32')
customer_data
```

```python
def merchant(Merchant_Name):
    return Merchant_Name.split(' ')[0].split('-')[0].split(',')[0]

customer_data['Merchandise'] = customer_data['Merchant Name'].apply(lambda x: merchant(x))
customer_data
```

## Question 1: Highest Generated Income by Category

```python
results = customer_data.groupby('Category').sum()
results
```

```python
category = ['Clothing', 'Cosmetics', 'Electronics', 'Market', 'Restaurant', 'Travel']
plt.bar(category, results['Transaction Amount'])
plt.xticks(category)
plt.ylabel('Highest Sales in USD ($)')
plt.xlabel('Category')
plt.title('Highest Generated Income by Category')
plt.show()
```

## Question 2: Best Month For Sales? 

```python
sales = customer_data.groupby('Month').sum()
sales
```

```python
months = range(1,11)

plt.bar(months, sales['Transaction Amount'])
plt.xticks(months)
plt.ylabel('Highest Sales in USD ($)')
plt.xlabel('Months')
plt.title('Best Month For Sales')
plt.show()
```
